2.Trend Analysis


1) Identify trends among the top YouTube streamers. Which categories are the most popular?


2) Is there a correlation between the number of subscribers and the number of likes or comments?

```python
#Finding of the popular categories
counts_category=df["Categories"].value_counts()
plt.figure(figsize=(18,10))
plt.title("List of Categories")
plt.xlabel("Categories")
plt.ylabel("Streamers")
plt.xticks(rotation=90)
sns.barplot(x=counts_category.index,y=counts_category.values,color="purple");
```

```python
#Finding of correlation between Subscribers and Likes
column1 = 'Subscribers'
column2 = 'Likes'
correlation = df[column1].corr(df[column2])
print("Correlation between Subscribers and Likes:",correlation)
```

```python
#Heatmap for the correlation of Subscribers and Likes
column1 = 'Subscribers'
column2 = 'Likes'

correlation_matrix = df[[column1, column2]].corr()
plt.figure(figsize=(8,6))
sns.heatmap(correlation_matrix, annot=True, cmap='viridis', fmt=".2f")
plt.title('Correlation Heatmap for Subscribers and Likes')
plt.show()
```

```python
#Finding of correlation between Subscribers and Comments
column1 = 'Subscribers'
column2 = 'Comments'
correlation = df[column1].corr(df[column2])
print("Correlation between Subscribers and Comments:",correlation)
#Heatmap for the correlation of Subscribers and Comments
correlation_matrix = df[[column1, column2]].corr()
plt.figure(figsize=(8,6))
sns.heatmap(correlation_matrix, annot=True, cmap='PuRd', fmt=".2f")
plt.title('Correlation Heatmap for Subscribers and Comments')
plt.show()
```

```python
#Scatter plot for Likes and Comments
sns.set_style("darkgrid")
plt.figure(figsize=(8,5))
sns.scatterplot(data=df, x='Likes', y='Comments',color='red')
plt.title('Scatter Plot of Likes vs Comments')
plt.xlabel('Likes')
plt.ylabel('Comments')
plt.show();
```

```python
#Cat plot for Subscribers and Comments
plt.figure(figsize=(8,5))
sns.catplot(data=df, x='Subscribers', y='Comments',kind='point',color='green')
plt.title('Cat Plot of Subscribers vs Comments')
plt.xlabel('Subscribers')
plt.ylabel('Comments')
plt.show();
```

4. Performance Metrics

1) Calculate and visualize the average number of subscribers, visits, likes, and comments.

2) Are there patterns or anomalies in these metrics?

```python
#Showing of the top10 subscribers along with their username (Tseries-1st)
X=df["Username"].head(10)
Y=df["Subscribers"].tail(10)
```

```python
#Piechart of showing the top10 subscribers
username=X
subscribers=Y
plt.pie(subscribers,labels=username,startangle=360)
plt.show();
```

```python
#Finding of the averages
avg_subscribers=df['Subscribers'].mean()
avg_likes=df['Likes'].mean()
avg_comments=df['Comments'].mean()
avg_visits=df['Visits'].mean()

print("Average Subscribers",avg_subscribers)
print("Average Likes",avg_likes)
print("Average Comments",avg_comments)
print("Average visits",avg_visits)

```

```python
#Linechart for the number of subscribers, likes ,commetns and visits
plt.figure(figsize=(12,8))
#Subscribers
plt.subplot(2,2,1)
plt.plot(df['Subscribers'],color='red')
plt.title("Subscribers")
#Likes
plt.subplot(2, 2, 2)
plt.plot( df['Likes'], color='black')
plt.title('Likes')
#Comments
plt.subplot(2,2,3)
plt.plot(df['Comments'],color='blue')
plt.title("Comments")
#Visits
plt.subplot(2,2,4)
plt.plot(df['Visits'],color='brown')
plt.title("Visits");
```

```python
#Finding of the average metrics of the dataset
metrics=['Subscribers','Likes','Comments','Visits']
average=[avg_subscribers,avg_likes,avg_comments,avg_visits]
plt.figure(figsize=(12,8))
plt.plot(metrics,average,'o-y')
sns.barplot(x=metrics, y=average, color='red')
plt.xlabel(metrics)
plt.ylabel(average)
plt.title("Average Performance Metrics")
plt.xticks(rotation=360)
plt.show();
```

5.Content Categories

1) Explore the distribution of content categories. Which categories have the highest number of
streamers?

2) Are there specific categories with exceptional performance metrics?

```python
#Showing of the number of streamers based on categories
category_counts=df['Categories'].value_counts()
plt.figure(figsize=(20,8))
sns.barplot(x=category_counts.index,y=category_counts.values,color='yellow')
plt.plot(category_counts.index, category_counts.values,'o--b')
plt.title("No of streamers based on content categories")
plt.xlabel("Content Categories")
plt.ylabel("No of streamers")
plt.xticks(rotation=90)
plt.show();
```

```python
#Showing of the category with highest number of streamers
#idxmax()-->shows the highest category rank
category_counts=df['Categories'].value_counts().idxmax()
print("Category with highest number of streamers:",category_counts)

```

```python
#Categories with exceptional performance
category_counts = df['Categories'].value_counts()
threshold = 40 #filtering of datas which is lower than 40
exceptional_categories = category_counts[category_counts > threshold]
print("Categories with Exceptional Performance:")
print(exceptional_categories)
```

```python
#Barplot for Categories with Exceptional Performance
plt.figure(figsize=(12, 8))
plt.plot(exceptional_categories.index,exceptional_categories.values,'o--r')
sns.barplot(x=exceptional_categories.index, y=exceptional_categories.values)
plt.title('Categories with Exceptional Performance')
plt.xlabel('Categories')
plt.ylabel('Subscribers Count')
plt.show()
```

6. Brands and Collaboration

1) Analyze whether streamers with high performance metrics receive more brand collaborations and
marketing campaigns.

```python
brands = df[['Categories','Comments', 'Likes', 'Visits']].sort_values(by = 'Likes' ,ascending=False)
top_brands = brands.head(5)
top_brands
```

```python
#Barplot for top 5 influenced brands
sns.barplot(data=top_brands, x='Categories', y='Likes',color='orange')
plt.title("Top 5 most Influenced Brands ")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

7. Benchmarking

1) Identify streamers with above-average performance in terms of subscribers, visits, likes, and comments.


2) Who are the top-performing content creators?

```python
#Streamers with above-average performance
df['Above_avgSubscribers'] = df['Subscribers'] > avg_subscribers
df['Above_avgVisits'] = df['Visits'] > avg_visits
df['Above_avgLikes'] = df['Likes'] > avg_likes
df['Above_avgComments'] = df['Comments'] > avg_comments
top_performing_streamers = df[df['Above_avgSubscribers'] & df['Above_avgVisits'] & df['Above_avgLikes'] & df['Above_avgComments']]
print(top_performing_streamers)
```

```python
#Top visiters of the channel
visit = df[['Username', 'Visits', 'Subscribers']].sort_values(by = 'Visits' ,ascending=False)
top_visit = visit.head(10)
top_visit
```

```python
#Barplot showing of the top10 content creators
plt.figure(figsize=(15,7))
sns.barplot(data=top_visit, x='Visits', y='Username', color='green')
plt.title("Top 10 content creators")
plt.xlabel("Visits")
plt.ylabel('Username')
plt.tight_layout()
plt.show()
```
